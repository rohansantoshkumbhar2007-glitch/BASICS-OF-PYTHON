```python
##python lists
thislist=["Mclaren","Redbull","Ferrari"]
print(thislist)
```

```python
#Allow Duplicates
thislist=["Mclaren","Redbull","Ferrari","Mclaren","Ferrari"]
print(thislist)
```

```python
#list Length
thislist=["Mclaren","Redbull","Ferrari"]
print(len(thislist))
```

```python
#A list with strings,integers and boolean values:
list1=["abc",34,True,"male"]
print(list1)
```

```python
#Access Items
thislist=["Mclaren","Redbull","Ferrari"]
print(thislist[1])
```

```python
#Reverse/negative Indexing
thislist=["Mclaren","Redbull","Ferrari"]
print(thislist[-1])
```

```python
#Slicing
thislist=["Mclaren","Redbull","Ferrari","Aston Martin","Mercedes","Audi"]
print(thislist[2:5])
print(thislist[-3:])
print(thislist[-4:])
```

```python
#Change Item Value
thislist=["Mclaren","Redbull","Ferrari"]
thislist[1]="Aston Martin"
print(thislist)
```

```python
#Append Items
thislist=["Mclaren","Redbull","Ferrari"]
thislist.append("Aston Martin")
print(thislist)
```

```python
#Insert Items
thislist=["Mclaren","Redbull","Ferrari"]
thislist.insert(2,"Aston Martin")
print(thislist)
```
